```python
#integrate pdf get cdf
##find derivative of cdf...slope of cdf ...get pdf
from matplotlib import pyplot as plt
```

```python
counts, bin_edges=np.histogram(iris_setosa['PL'],bins=10)
```

```python
print(counts)
print(bin_edges)
```

```python
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
plt.style.use('fivethirtyeight')
counts, bin_edges=np.histogram(iris_setosa['PL'],bins=10)
pdf=counts/sum(counts)

cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts, bin_edges=np.histogram(iris_versicolor['PL'],bins=10)
pdf=counts/sum(counts)

cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts, bin_edges=np.histogram(iris_virginica['PL'],bins=10)
pdf=counts/sum(counts)

cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
#cdf helps better analysis at segregating data than pdfs evidently
```
